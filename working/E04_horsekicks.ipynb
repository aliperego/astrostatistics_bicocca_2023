```python
import numpy as np
from matplotlib import pyplot as plt
import scipy
from scipy.stats import poisson

plt.rcParams['figure.figsize'] = [5, 5]
%config InlineBackend.figure_format='retina'
```

There are 200 army corps.
- 109 corps had no death
- 65 corps had 1 death -> 65x1=65 deaths
- 22 corps had 2 death -> 22x2=44 deaths
- 3 corps had 3 death -> 3x3=9 deaths
- 1 corps had 4 deaths -> 1x4=4 deaths

Total number of deaths = 122\
Total number of corps = 200

```python
deaths = np.array([0,1,2,3,4])
corps = np.array([109,65,22,3,1])

mu = np.average(deaths,weights=corps)
print('The mean is : ',mu)
```

```python
fig,axs = plt.subplots(1,2,figsize=(12,5))
axs[0].bar(deaths, corps)
axs[0].set_xlabel('Deaths')
axs[0].set_ylabel('Number of corps')

dist = scipy.stats.poisson(mu)
axs[1].bar(deaths,corps/200, alpha=0.5)
axs[1].plot(deaths,dist.pmf(deaths), '-o',label = 'Poisson distribution with $\mu=%.2f$' %mu)
axs[1].set_xlabel('Deaths')
axs[1].set_ylabel('Percentage of corps')

axs[1].legend()
plt.show()
```
