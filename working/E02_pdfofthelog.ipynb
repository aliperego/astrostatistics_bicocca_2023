```python
import numpy as np
import matplotlib.pyplot as plt
import random
import astroML
from scipy import stats
from astropy.visualization import hist
plt.rcParams['figure.figsize'] = [5, 5]
%config InlineBackend.figure_format='retina'
```

```python
N=10000
x = np.random.uniform(0.1, 10, N)
plt.hist(x, bins=30, density=True)
plt.xlabel('x')
plt.show()
```

- $x$ has a uniform distribution between $0.1$ and $10$, so $p(x) = \frac{1}{10-0.1} = \frac{1}{9.9}$.

- $y=\log_{10} (x)$ and $x=10^{y}$ , so $p(y) = \left|\frac{dx}{dy}\right|p(x) = 10^{y} (\ln{10}) (\frac{1}{9.9})$


```python
y = np.log10(x)
plt.hist(y, bins=50, label='histogram of y', density=True);

y_grid = np.linspace(-1, 1, 10000)
p = 10**y_grid * np.log(10) * (1/9.9)
plt.plot(y_grid, p, lw=2, color="C1", label='p(y)')

plt.legend()
plt.show()
```

```python
mean_x = np.mean(x)
mean_y = np.mean(y)

print("The log of the mean of x is ", np.log10(mean_x))
print("The mean of y is ", mean_y)
print("They are different!")
```

```python
med_x = np.median(x)
med_y = np.median(y)

print("The log of the median of x is ", np.log10(med_x))
print("The median of y is ", med_y)
print("They are the same!")
```
